```python
import numpy as np
import matplotlib.pyplot as plt

def cost_function(x):
    return (x + 3)**2

def gradient(x):
    return 2 * (x + 3)

def gradient_descent(starting_point, learning_rate, precision, max_iterations):
    x_values = []
    cost_values = []

    x = starting_point
    step_size = 1
    iterations = 0

    while step_size > precision and iterations < max_iterations:
        prev = x
        x = x - learning_rate * gradient(prev)
        step_size = abs(prev - x)
        iterations += 1

        cost = cost_function(x)
        x_values.append(x)
        cost_values.append(cost)

        print(f"Iteration {iterations}: x = {x}, Cost = {cost}, Step Size = {step_size}")

    return x_values, cost_values

# Parameters
starting_point = 2
learning_rate = 0.1
precision = 0.00001
max_iterations = 100

# Run gradient descent
x_values, cost_values = gradient_descent(starting_point, learning_rate, precision, max_iterations)

# Visualize the cost function and the path taken by gradient descent
x_range = np.linspace(-10, 5, 100)
cost_values_range = cost_function(x_range)

plt.plot(x_range, cost_values_range, label='Cost Function')
plt.scatter(x_values, cost_values, color='red', label='Gradient Descent Path')
plt.title('Gradient Descent Visualization')
plt.xlabel('x')
plt.ylabel('Cost')
plt.legend()
plt.show()

# Display the final local minimum
minima = x_values[-1]
minima_value = cost_values[-1]

print(f"Local minima found at x = {minima} with a cost value of {minima_value}")
```
